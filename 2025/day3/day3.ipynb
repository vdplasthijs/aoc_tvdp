```python
input_text = au.read_txt_file_lines()
n_rows = len(input_text)
n_cols = len(input_text[0])
for ii in range(1, n_rows):
    assert len(input_text[ii]) == n_cols, f'row {ii} has {len(input_text[ii])} cols, not {n_cols}'
print(f'input has {n_rows} rows and {n_cols} cols')
```

## Part 1

```python
total = 0 

for b in input_text:
    for k in range(9, 0, -1):
        if str(k) in b[:-1]:
            break ## max 
    trail = b.split(str(k))
    assert len(trail) > 1
    trail = str(k).join(trail[1:])
    for l in range(9, 0, -1):
        if str(l) in trail:
            num = int(str(k) + str(l))
            total += num 
            break

total

    
```

## Part 2

```python
## same approach but iterative .??? 

n_nums = 12
total = 0 

for b in input_text:
    trail = b
    best_nums = []
    for n_trail in range(n_nums - 1, -1, -1):  # iterate through n digits left
        for k in range(9, 0, -1):
            if str(k) in trail[:(len(trail) - n_trail)]:  # get first max
                best_nums.append(str(k))
                break ## max 
        new_trail = trail.split(str(k))
        # print(n_trail, trail[:(len(trail) - n_trail)], k)
        assert len(new_trail) > 1
        trail = str(k).join(new_trail[1:])
        assert len(trail) >= n_trail, (len(trail), n_trail)
    
    jolt = int(''.join(best_nums))
    assert len(str(jolt)) == n_nums, len(str(jolt))
    total += jolt

total

    
```
