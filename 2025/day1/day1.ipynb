```python
n_zeros = 0
dial = 50 

for add in additions:
    prev_dial = dial
    # print(dial, add, n_zeros)
    if add > 0:
        dial += add 
        if dial >= 100:
            n_100 = dial // 100 
        else:
            n_100 = 0
    elif add < 0:
        dial += add
        if dial <= 0:
            if dial == 0:
                n_100 = -1
            else:
                n_100 = dial // 100
                if dial % 100 == 0:
                    n_100 -= 1
            if prev_dial == 0:
                n_100 += 1
            n_100 = -1 * n_100
        else:
            n_100 = 0
    else:
        assert False, 'unexpected'
    n_zeros += n_100
    dial = dial % 100
    # print('n_100', n_100)

print(n_zeros)
```
