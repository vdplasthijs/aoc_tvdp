```python
input_text = au.read_txt_file_lines()
n_rows = len(input_text)
n_cols = len(input_text[0])
# for ii in range(1, n_rows):
#     assert len(input_text[ii]) == n_cols, f'row {ii} has {len(input_text[ii])} cols, not {n_cols}'
print(f'input has {n_rows} rows and {n_cols} cols')
```

```python
ranges, numbers = [], []
add_range = True
for line in input_text:
    if line == '':
        print('switch')
        add_range = False
        continue
    if add_range:
        ranges.append([int(k) for k in line.split('-')])
    else:
        numbers.append(int(line))
```

```python
all_ranges = [range(r[0], r[1] + 1) for r in ranges]
total = 0
for n in numbers:
    for r in all_ranges:
        if n in r:
            total += 1
            break 

total
```
