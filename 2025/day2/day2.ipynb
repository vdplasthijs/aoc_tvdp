```python
input_text = au.read_txt_file_lines()
n_rows = len(input_text)
n_cols = len(input_text[0])
for ii in range(1, n_rows):
    assert len(input_text[ii]) == n_cols, f'row {ii} has {len(input_text[ii])} cols, not {n_cols}'
print(f'input has {n_rows} rows and {n_cols} cols')
```

```python
ranges = [k.split('-') for k in input_text[0].split(',')]
ranges = [(int(k[0]), int(k[1])) for k in ranges]
ranges
```

```python
total = 0

for r in ranges:
    r0, r1 = r
    n_digits = [l for l in range(len(str(r0)), len(str(r1)) + 1)]
    n_digits = [n for n in n_digits if n % 2 == 0]  # only even num digits can be possible
    # print(r, n_digits)
    for i_n, n in enumerate(n_digits):
        n_half = n // 2

        ## Finding start and end of range (can be replaced by 100 and 999 (eg for 3) I think as it's quick anyway)
        if n == len(str(r0)): 
            d_start = str(r0)[:n_half]
            if int(d_start + d_start) < r0:
                d_start = str(int(d_start) + 1)
        else:
            d_start = '1' + '0' * (n_half - 1)

        if n == len(str(r1)):
            d_end = str(r1)[:n_half]
        else:
            d_end = '9' * n_half
        # print(r0, r1, n, d_start, d_end)

        ## Add 
        for d in range(int(d_start), int(d_end) + 1):
            num = int(str(d) * 2)
            if num in range(r0, r1 + 1):  # check for validity, eg if start is 2233, then 2222 won't go. Probably only necessary for early numbers until >r0
                total += num 

total


            
```

```python
total = 0
set_nums = set()

for r in ranges:
    r0, r1 = r
    n_digits = [l for l in range(len(str(r0)), len(str(r1)) + 1)]
    # print(r, n_digits)
    for i_n, n_total in enumerate(n_digits):
        n_total_half = n_total // 2
        for n in range(1, n_total_half  + 1):
            if n_total % n != 0:  # get all nums that mod=0 so their combis can fit. 
                continue 
            n_combis = int(n_total / n)
    
            num_start = int('1' + '0' * (n - 1))  #start and end at 100/999 (eg for 3)
            for num in range(num_start, int('9' * n) + 1):
                full_num = int(str(num) * n_combis)
                if full_num < r0:
                    continue 
                elif full_num > r1:  # end of range, break loop
                    break 
                else:
                    if full_num not in set_nums:  # check for eg '2' x 4 vs '22' x 2
                        set_nums.add(full_num)
                        total += full_num

total

```
