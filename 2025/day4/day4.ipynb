```python
input_text = au.read_txt_file_lines()
n_rows = len(input_text)
n_cols = len(input_text[0])
for ii in range(1, n_rows):
    assert len(input_text[ii]) == n_cols, f'row {ii} has {len(input_text[ii])} cols, not {n_cols}'
print(f'input has {n_rows} rows and {n_cols} cols')
```

```python
def convert(x):
    if x == '.':
        return 0
    elif x == '@':
        return 1
    raise ValueError

maze = np.array([[convert(x) for x in y] for y in input_text])
n = maze.shape[0]

def max_bound(x, n=n):
    return min(x + 2, n)

def min_bound(x):
    return max(x - 1, 0)

total = 0

for r in range(n):
    for c in range(n):
        if maze[r, c] == 0:
            continue 
        r_min, r_max = min_bound(r), max_bound(r)
        c_min, c_max = min_bound(c), max_bound(c)

        square = maze[r_min:r_max, c_min:c_max]
        if np.sum(square) <= 4:
            total += 1

total
```
