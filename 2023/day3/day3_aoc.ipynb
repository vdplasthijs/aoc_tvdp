```python
%load_ext autoreload
%autoreload 2

import numpy as np
import os, sys 
sys.path.append('..')

import aoc_utils as au
```

```python
input_text = au.read_txt_file_lines()
input_text[:4]
```

```python
all_chars = set() 

for ii in range(len(input_text)):
    all_chars = all_chars.union(set(input_text[ii]))
    if ii == 0:
        len_row = len(input_text[ii])
    else:
        assert len_row == len(input_text[ii])
all_symbols = all_chars - set(['.']) - set([x if x.isdigit() else '' for x in all_chars])

id_num = 0
dict_all_numbers = {}
for ii in range(len(input_text)):
    jj = 0 
    while jj < len(input_text[ii]):
        if input_text[ii][jj].isdigit():
            k = 1
            full_number = input_text[ii][jj]
            while jj + k < len(input_text[ii]) and input_text[ii][jj + k].isdigit():
                full_number += input_text[ii][jj + k]
                k += 1
            dict_all_numbers[id_num] = {'row': ii, 'col_start': jj, 'col_end': jj + k - 1, 'number': int(full_number)}
            id_num += 1
            jj += k
        else:
            jj += 1

print(dict_all_numbers)
```

```python
list_numbers_adjacent_to_symbols = []
for i_num, num_info in dict_all_numbers.items():
    first_row_check = np.maximum(num_info['row'] - 1, 0)
    last_row_check = np.minimum(num_info['row'] + 1, len(input_text) - 1)
    first_col_check = np.maximum(num_info['col_start'] - 1, 0)
    last_col_check = np.minimum(num_info['col_end'] + 1, len(input_text[0]) - 1)

    bool_append_num = False
    for ii in range(first_row_check, last_row_check + 1):
        for jj in range(first_col_check, last_col_check + 1):
            if input_text[ii][jj] in all_symbols:
                bool_append_num = True 
                break
    if bool_append_num:
        list_numbers_adjacent_to_symbols.append(num_info['number'])

# print(list_numbers_adjacent_to_symbols)
print(np.sum(list_numbers_adjacent_to_symbols))
```

```python
## cheeky copy-paste from part 1, now just save numbers that are adjacent to asterisks and store per asterisk id 

dict_asterisks_adjacent_nums = {}

for i_num, num_info in dict_all_numbers.items():
    first_row_check = np.maximum(num_info['row'] - 1, 0)
    last_row_check = np.minimum(num_info['row'] + 1, len(input_text) - 1)
    first_col_check = np.maximum(num_info['col_start'] - 1, 0)
    last_col_check = np.minimum(num_info['col_end'] + 1, len(input_text[0]) - 1)

    for ii in range(first_row_check, last_row_check + 1):
        for jj in range(first_col_check, last_col_check + 1):
            if input_text[ii][jj] == '*':
                ind_ast = (ii, jj)
                if ind_ast not in dict_asterisks_adjacent_nums:
                    dict_asterisks_adjacent_nums[ind_ast] = []
                dict_asterisks_adjacent_nums[ind_ast].append(num_info['number'])
print(dict_asterisks_adjacent_nums)
```

```python
sum_prods = 0 
for ind_ast, list_nums in dict_asterisks_adjacent_nums.items():
    if len(list_nums) == 2:
        sum_prods += np.prod(list_nums)

print(sum_prods)
```
